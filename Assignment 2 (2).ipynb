```python
# Importing necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
data_url = "https://www.kaggle.com/uciml/german-credit"
credit_data = pd.read_csv("german_credit_data.csv")
```

```python
# Display basic statistics of the dataset
print("Basic Statistics of the Dataset:")
print(credit_data.describe())
```

```python
# Display correlation matrix for numeric columns
numeric_data = credit_data.select_dtypes(include=['int64', 'float64'])
print("\nCorrelation Matrix:")
print(numeric_data.corr())
```

```python
# Display basic statistics of the dataset
print("Basic Statistics of the Dataset:")
print(credit_data.describe())
```

```python
# Choose two variables for scatter plot
x_variable = 'Credit amount'
y_variable = 'Duration'

# Create scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(credit_data[x_variable], credit_data[y_variable], alpha=0.5, color='blue')
plt.title('Scatter Plot of Credit Amount vs. Duration')
plt.xlabel('Credit Amount (in Euros)')
plt.ylabel('Duration (in Months)')
plt.grid(True)

# Add regression line
sns.regplot(x=x_variable, y=y_variable, data=credit_data, scatter=False, color='red')

# Add legend
plt.legend(['Regression Line'])

# Save plot as image
plt.savefig('scatter_plot.png', dpi=200)

# Show plot
plt.show()
```

```python
# Choose a categorical variable for bar graph
categorical_variable = 'Purpose'

# Calculate value counts for the categorical variable
category_counts = credit_data[categorical_variable].value_counts()

# Create bar graph
plt.figure(figsize=(10, 6))
category_counts.plot(kind='bar', color='skyblue', edgecolor='black')  # Adding edgecolor for better visibility of bars
plt.title('Distribution of Credit Purposes')
plt.xlabel('Purpose')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y')

# Add data labels
for i, count in enumerate(category_counts):
    plt.text(i, count, str(count), ha='center', va='bottom')

plt.tight_layout()  # Adjust layout to prevent overlap of labels

# Save plot as image
plt.savefig('bar_plot.png', dpi=200)
plt.show()
```

```python
def create_heatmap(data):
    # Compute correlation matrix
    corr_matrix = data.corr()

    # Create heatmap
    plt.figure(figsize=(10, 8))
    sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
    plt.title('Correlation Heatmap')
    plt.show()

# Display correlation matrix for numeric columns
numeric_data = credit_data.select_dtypes(include=['int64', 'float64'])

# Create heatmap
create_heatmap(numeric_data)
```

```python
def create_pie_chart(dataframe, categorical_variable):
    """
    Create a pie chart to visualize the distribution of a categorical variable.

    Args:
    - dataframe: Pandas DataFrame containing the dataset
    - categorical_variable: Name of the categorical variable to be visualized

    Returns:
    - None
    """
    # Calculate value counts for the categorical variable
    category_counts = dataframe[categorical_variable].value_counts()

    # Create pie chart
    plt.figure(figsize=(8, 8))
    plt.pie(category_counts, labels=category_counts.index, autopct='%1.1f%%', startangle=140)
    plt.title('Distribution of ' + categorical_variable)
    plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle
    plt.savefig('pie_plot.png', dpi=200)
    plt.show()

# Choose a categorical variable for pie chart
categorical_variable = 'Housing'

# Create pie chart
create_pie_chart(credit_data, categorical_variable)
```
